```python
# Install necessary libraries
# This should be in a requirements.txt file, not executed within the notebook
# !pip install fastai fastbook voila

# Import the required modules
from fastai.vision.all import *
from fastai.vision.widgets import *

# Assuming export.pkl is in the same directory as your notebook when deployed
path = Path('.')
learn_inf = load_learner(path/'export.pkl')

# Upload button
btn_upload = FileUpload()

# Output widget
out_pl = Output()

# Label for predictions
lbl_pred = Label()

# Button for classifications
btn_run = Button(description='Classify')

# Event handler for classification
def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128, 128))
    pred, pred_idx, probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)

# Assemble widgets
widgets = VBox([Label('Select your panel!'), btn_upload, btn_run, out_pl, lbl_pred])
display(widgets)
```
